# Exploratory Data Analysis (EDA)

## Bank Customer Churn Analysis

```python

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
%matplotlib inline
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
# This section shows the number of rows and columns present in the dataset.
df.shape
```

```python
df.columns
# Let's examine all the features available in the dataset.
```

```python

 # This visualization shows the number of customers who stayed and who left the bank.
sns.countplot(data=df, x="Exited")
plt.title("Customer Churn Distribution")
plt.show()

```

4. Numerical Features Distribution
- CreditScore
- Age
- Balance
- EstimatedSalary

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()
```

5. Categorical Features
- Geography
- Gender
- HasCrCard
- IsActiveMember

```python
sns.countplot(data=df, x="Gender")
plt.show()
```

```python
plt.figure(figsize=(10,7))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```

```python
sns.boxplot(data=df, x="Exited", y="Age")
plt.show()
```

```python
sns.boxplot(data=df, x="Exited", y="Balance")
plt.show()
```

```python
sns.boxplot(data=df, x="Exited", y="EstimatedSalary")
plt.show()
```

```python
sns.pairplot(
    df[
        [
            "CreditScore",
            "Age",
            "Balance",
            "EstimatedSalary",
            "Exited"
        ]
    ],
    hue="Exited"
)
```
