# Bank Customer Churn Prediction - Model Saving

## Objective

In this notebook, we will:

- Load the processed dataset
- Train the best performing Random Forest model
- Save the trained model
- Save the StandardScaler
- Save feature column names
- Load the saved files
- Verify that everything works correctly

This notebook prepares the project for deployment.

```python

import os
import joblib
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("../data/processed/churn_processed.csv")

df.head()
```

```python
X = df.drop("Exited", axis=1)
y = df["Exited"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train_scaled, y_train)

print("✅ Random Forest model trained successfully!")
```

```python
os.makedirs("../models", exist_ok=True)
```

```python
joblib.dump(rf_model, "../models/random_forest_model.pkl")

print("✅ Model saved successfully!")
```

```python
joblib.dump(scaler, "../models/scaler.pkl")

print("✅ Scaler saved successfully!")
```

```python
feature_columns = X.columns.tolist()

joblib.dump(feature_columns, "../models/feature_columns.pkl")

print("✅ Feature columns saved successfully!")
```

```python
loaded_model = joblib.load("../models/random_forest_model.pkl")
loaded_scaler = joblib.load("../models/scaler.pkl")
loaded_features = joblib.load("../models/feature_columns.pkl")

print(type(loaded_model))
print(type(loaded_scaler))
print(loaded_features)
```

```python
sample = X.iloc[[0]]

sample_scaled = loaded_scaler.transform(sample)

prediction = loaded_model.predict(sample_scaled)

probability = loaded_model.predict_proba(sample_scaled)

print("Prediction:", prediction[0])
print("Probability:", probability)
```
